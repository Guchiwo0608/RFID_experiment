```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# 定数の設定
mu = 2 * np.pi - 0.3  # 平均
sigma = 1 / 3         # 標準偏差
x_min, x_max = 0, 2*np.pi   # xの範囲

# xの値を生成
x = np.linspace(x_min, x_max, 500)

# 正規分布の確率密度関数
y = norm.pdf(x, mu, sigma) + norm.pdf(x, mu-2*np.pi, sigma)

# # xが2π以上の値を2π剰余演算
# x_mod = np.mod(x, 2 * np.pi)
# y_mod = norm.pdf(x_mod, mu, sigma)

# プロット
fig = plt.figure(figsize=(3.5, 2.5))
plt.plot(x, y, label=f'N({mu:.2f}, {sigma**2:.2f})', color='blue')
plt.vlines(mu, -1, 10)
plt.ylim(0, 1.5)
plt.xlim(0, 2*np.pi)
plt.xlabel("Phase[rad]")
plt.ylabel("probabilistic density")
plt.fill_between(x, y, where=(0<=x)&(x<=1), fc="red")


# グラフを表示
plt.show()
```
